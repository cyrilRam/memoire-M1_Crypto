## Indice de MarchÃ©

```python
#crÃ©ation Indice de marchÃ© = protef equipondÃ©rÃ© des cryptos sÃ©lÃ©ctionnÃ©s
datIndice=pd.DataFrame(columns=["Indice"],index=dataCoursCrypt.index)
datIndice["Indice"]=dataRendCrypt.mean(axis=1)
DataSet['Indice MarchÃ©']=datIndice
```

## Size

```python
#crÃ©ation du facteur Size : 

# CrÃ©ation des listes gros capi et small capi en fonction de yahoo finance
lstBigCap=["Bitcoin","Ethereum","Tether","Xrp","Cardano","Binance Coin"]
lstSmallCap=["Bitcoin Cash","Chainlink","Litecoin","Decentraland","Tron","Dogecoin"]

#crÃ©ation du facteur size
datCap=pd.DataFrame(columns=["BigCaps","SmallCaps"],index=dataCoursCrypt.index)
datCap["BigCaps"]=dataRendCrypt[lstBigCap].mean(axis=1)
datCap["SmallCaps"]=dataRendCrypt[lstSmallCap].mean(axis=1)
datCap["Size"]=datCap[["BigCaps","SmallCaps"]].apply(lambda r:r[1]-r[0],axis=1)
DataSet['Size']=datCap.loc[:,"Size"]
```

## Momentum

```python
#crÃ©ation du facteur Momentum :  

# crÃ©ation periode de rÃ©fernce pour le calcul des Winners et des loosers
DataReturnCrypt=dataRendCrypt.loc[date(2018,1,3):"2019"]

# calcul rendement moyen avec une fenetre roulante de 14 jours 
# => rÃ©sultat chaqe ligne correspondt Ã  la moyenne des 14 jours precedents
DataReturnCrypt=DataReturnCrypt.rolling(window=14).mean()


# calcul de la perf moyenne des rendements sur la totalitÃ© de l'historique
datAvarageReturn=DataReturnCrypt.loc[date(2018,1,16):"2019"].mean(axis=0)

#tri en fonction de la perf sur la pÃ©riode de reference
datAvarageReturn=datAvarageReturn.sort_values(axis=0, ascending=True)


# crÃ©ation des liste winners et losser en fonction du tri
lstLosers=list(datAvarageReturn.index[0:5])
lstWinners=list(datAvarageReturn.index[6:])

#crÃ©ation du facteur momentum
datMom=pd.DataFrame(columns=["Winners","Losers"],index=dataCoursCrypt.index)
datMom["Winners"]=dataRendCrypt[lstWinners].mean(axis=1)
datMom["Losers"]=dataRendCrypt[lstLosers].mean(axis=1)
datMom["Momentum"]=datMom[["Winners","Losers"]].apply(lambda r:r[1]-r[0],axis=1)
DataSet['Momentum']=datMom.loc[:,"Momentum"]
```

## VolatilitÃ©

```python
#crÃ©ation du facteur volatilitÃ© des crypto : 

#On considere les rendements
DataReturnCrypt=dataRendCrypt

# calcul rendement moyen tous les 14 jours avec une fenetre roulante 
DataReturnCrypt=DataReturnCrypt.rolling(window=14).mean()

# calcul de l'ecart-type  sur les rendements obtenus avec la fenetre roulante
datstd=DataReturnCrypt.loc[date(2018,1,16):"2019"].std(axis=0)

#tri en fonction de l'ecart-type sur la pÃ©riode de reference
datstd=datstd.sort_values(axis=0, ascending=True)


# crÃ©ation des liste vol Ã©lÃ©vÃ© et vol faible en fonction du tri
lstLowVol=list(datstd.index[0:5])
lstHighVol=list(datstd.index[6:])


#crÃ©ation du facteur volatilitÃ© : vol stable moins vol Ã©lÃ©vÃ©
datVol=pd.DataFrame(columns=["HighVol","LowVol"],index=dataCoursCrypt.index)
datVol["HighVol"]=dataRendCrypt[lstHighVol].mean(axis=1)
datVol["LowVol"]=dataRendCrypt[lstLowVol].mean(axis=1)
datVol["VolatilitÃ©"]=datVol[["HighVol","LowVol"]].apply(lambda r:r[1]-r[0],axis=1)
DataSet['VolatilitÃ©']=datVol.loc[:,"VolatilitÃ©"]
```

# Ajout des autres Facteurs

```python
#creation fonction pour recuperer les donnÃ©es et modifier les nan du we
def AjoutFacteur(Data,Ticker,Name,dateStart,rend):
    df=yf.download(['BTC-USD',Ticker],start=dateStart,end=date.today(),interval = "1d")
    
    if rend==False:
        dataFact=df['Adj Close'].pct_change()
    else:
        dataFact=df['Adj Close']
    
    #on remplace les NAN par la derniÃ¨re valeur connue 
    dataFact[Ticker]=dataFact[Ticker].fillna(method='ffill')
    dataFact=dataFact.drop('BTC-USD',axis=1)

    Data[Name]=dataFact
    return Data
```
